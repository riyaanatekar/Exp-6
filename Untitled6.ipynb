```python
#9 salary calculations with allowances
'''problem statement:
write a python program to calculate the gross salary of

if basics <= 10000: HRA = 20% , DA = 80%
if basics <= 20000: HRA = 25% , DA = 90%
OTHERWISE HRA = 30% , DA = 95%'''

basic = float(input("Enter basic salary: "))

if basic <= 10000:
    hra = basic * 0.20
    da = basic * 0.80
elif basic <= 20000:
    hra = basic * 0.25
    da = basic * 0.90
else:
    hra = basic * 0.30
    da = basic * 0.95

gross_salary = basic + hra + da
print("Gross Salary:", gross_salary)
```

```python
#10. Write a program to calculate income tax based on annual income:
# up to 2,50,000: no tax
# 2,50,001 - 5,00,000: 5%
# 5,00,001 - 10,00,000: 20%
# above 10,00,000: 30%

try:
    income = float(input("Enter annual income: "))

    if income <= 250000:
        tax = 0
    elif income <= 500000:
        tax = (income - 250000) * 0.05
    elif income <= 1000000:
        tax = (250000 * 0.05) + (income - 500000) * 0.20
    else:
        tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000) * 0.30

    print("Income Tax:", tax)
except ValueError:
    print("Error: Please enter a valid numeric value.")
```
